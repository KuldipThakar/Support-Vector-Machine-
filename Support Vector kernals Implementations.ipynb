```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Circle 1 (radius = 5)
x = np.linspace(-5, 5, 200)  # Ensure full range for roundness
y = np.sqrt(5**2 - x**2)
x = np.hstack([x, x])  # Repeat x for upper and lower halves
y = np.hstack([y, -y])  # Reflect y to get both halves
```
