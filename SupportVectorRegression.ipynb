## feature eccoding[label encoding , onehot encoding ]

### Feature encoding for those who has binary data only two catogories , whenever we have two catagories we can replafce it with 1 and 0 , like smoker non smoker can be replaced with 1 and 0 

### for more then two catagories we apply onehot encoding 



```python
df.columns
```

```python
#independent feature 

x = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]

#dependent feature 

y = df[['total_bill']]
```

```python
#train test split 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size= 0.24, random_state= 10)
```

```python
X_train.head()
```

#feature encoding  (label encoding and onehot encoding )

#before doing feature engineering always train test split the data , otherwise data leakage may happen

#model should have no information about the test data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex']= le1.fit_transform(X_train["sex"])
X_train["smoker"]= le2.fit_transform(X_train["smoker"])
X_train["time"]= le3.fit_transform(X_train["time"])
```

```python
X_train.head()
```

```python
X_test['sex']= le1.transform(X_test["sex"])
X_test["smoker"]= le2.transform(X_test["smoker"])
X_test["time"]= le3.transform(X_test["time"])
```

```python
## One hot encoding --- column transformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('OneHot',OneHotEncoder(drop='first'),[3])],remainder='passthrough')
```

```python
import sys
import numpy 

numpy.set_printoptions(threshold=sys.maxsize)

X_train= ct.fit_transform(X_train)
```

```python
X_test= ct.transform(X_test)
```

```python
X_test
```

```python
###SVR: Support vector Regressor 

from sklearn.svm import SVR
svr = SVR()
```

```python
y_train
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
### Hyperparameter tuning using gridserachcv

from sklearn.model_selection import GridSearchCV
```

```python
##defining paramgrid

paramgrid = { 'C':[0.1,1.0,10,100,1000],
             'gamma' : [1,0.1,0.01,0.001,0.0001],
             'kernel':['rbf']}
```

```python
grid= GridSearchCV(SVR(),paramgrid,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_prediction = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error

print(r2_score(y_test,grid_prediction))
print(mean_absolute_error(y_test,grid_prediction))
```
